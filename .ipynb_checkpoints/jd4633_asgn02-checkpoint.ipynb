# Initial Setup

First, we will create the Perceptron class, using the code provided from the book. The one change we make is to add a "printEvery" parameter to the fit method. This will let us see the change in weights with each iteration if we want, by setting the parameter to a number, which tells how many epochs to run before printing out the weights.

```python
import numpy as np


class Perceptron(object):
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.

    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    errors_ : list
      Number of misclassifications (updates) in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y, printEvery=0):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.errors_ = []
        if (printEvery > 0):
            print("0", ": " , self.w_)

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            if (printEvery > 0 and ((_+1) % printEvery == 0)):
                print(_+1, ": " , self.w_)                
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

Next, we will create the Adaline class, again from code provided by the book. We again add the "printEvery" parameter to the fit method.

```python
class AdalineGD(object):
    """ADAptive LInear NEuron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.


    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    cost_ : list
      Sum-of-squares cost function value in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y, printEvery=0):
        """ Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.cost_ = []
        if (printEvery > 0):
            print("0", ": " , self.w_)

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            # Please note that the "activation" method has no effect
            # in the code since it is simply an identity function. We
            # could write `output = self.net_input(X)` directly instead.
            # The purpose of the activation is more conceptual, i.e.,  
            # in the case of logistic regression (as we will see later), 
            # we could change it to
            # a sigmoid function to implement a logistic regression classifier.
            output = self.activation(net_input)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
            if (printEvery > 0 and ((i+1) % printEvery == 0)):
                print(i+1, ": " , self.w_)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """Compute linear activation"""
        return X

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```

Now we define the function "plot_decision_regions" using the code provided from the book, which will help us to view the decision regions defined by the models.

```python
from matplotlib.colors import ListedColormap


def plot_decision_regions(X, y, classifier, resolution=0.02):

    # setup marker generator and color map
    markers = ('o', 's', 'x', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
```

Next we download the IRIS dataset from the public website and store it in a pandas dataframe. We print out 5 rows of each Iris type from the dataset to validate it loaded properly.

```python
import pandas as pd

df = pd.read_csv('https://archive.ics.uci.edu/ml/'
        'machine-learning-databases/iris/iris.data', header=None)
print(df.head())
print(df.iloc[50:54])
print(df.tail())
```

# Linearly Separable Classes
## 2 Features
Now, we set our feature matrix (X) and target vector (Y). The instructions are to pick two species of Iris, and 2 of the four features, such that the data sets are linearly separable. We'll choose "setosa" (the first 50 records, whose target value is 0) and "versicolor" (the next 50 records, whose target value is 1). For the features, we'll choose Sepal Length (column 0) and Petal Length (column 2). So, we'll take only the first 100 rows from the two feature and target data set, and only columns 0 and 2 from the feature set. For the target data set, we'll convert "Iris-setosa" to "1" and "Iris-versicolor" to "-1", so that setosa is a "success" for the Perceptron and versicolor is a "fail".

In addition, we use the "test_train_split" method from scikit_learn to divide our data into a training set (70%) and testing set (30%). This has additional benefits, such as shuffling the data sets so that the setosa records are interleved with the versicolor records.

```python
from sklearn.model_selection import train_test_split

# extract sepal length and petal length
#0=sepal length, 1=sepal width, 2=petal length, 3=petal width
X1 = df.iloc[0:100, [0, 2]].values

# select setosa and versicolor
y1 = df.iloc[0:100, 4].values
y1 = np.where(y1 == 'Iris-setosa', 1, -1)

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=1, stratify=y1)
```

Next we will plot the data using matplotlib, to show that the data is linearly separable.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# plot data
plt.scatter(X1[:50, 0], X1[:50, 1],
            color='red', marker='o', label='setosa')
plt.scatter(X1[50:100, 0], X1[50:100, 1],
            color='blue', marker='x', label='versicolor')

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

plt.show()
```

## 2 Features - Perceptron
Now we'll apply our Perceptron model to the data. We'll set eta to 0.1, the number of iterations to 10, and set printEvery to 1 to print the weights after each iteration.
We can see that the Perceptron model stabilizes after the first iteration, making 7 updates during the first epoch. Viewing the weights, we can see that w1 is positive; this means that high sepal length compared to petal length is positively related to being setosa. w2 is negative, which means that high petal length compared to sepal length, is negatively related to being setosa.  

```python
ppn1 = Perceptron(eta=0.1, n_iter=10)

ppn1.fit(X1_train, y1_train, printEvery=1)

plt.plot(range(1, len(ppn1.errors_) + 1), ppn1.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')

plt.show()
```

Below we view the decision region created by the Perceptron. While Perceptron is not guaranteed to produce an optimal line, the line actually divides the classes pretty evenly.

```python
plot_decision_regions(X1_train, y1_train, classifier=ppn1)
plt.title('Perceptron')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Now, we test the trained model on the feature set. We find that our model predicts the test set with 100% accuracy.

```python
y1_ppn_pred = ppn1.predict(X1_test)
ppn_misses1 = (y1_test != y1_ppn_pred).sum()
ppn_accuracy1 = (1 - ppn_misses1/y1_ppn_pred.size) * 100
print(f" model made {ppn_misses1} misses out of {y1_ppn_pred.size} tests for an accuracy of {ppn_accuracy1}%")
```

## 2 Features - Adaline
Next we apply the Adaline model to the data. Through trial and error, we find that 0.0003 is a good learning rate. The sum-squared error is pretty close to leveling off after 100 epochs at that learning rate.

```python
ax = plt.axes()

ada1 = AdalineGD(n_iter=100, eta=0.0003).fit(X1_train, y1_train)
plt.plot(range(1, len(ada1.cost_) + 1), ada1.cost_, marker='o')
ax.set_xlabel('Epochs')
ax.set_ylabel('Sum-squared-error')
ax.set_title('Adaline')

plt.show()
```

Next we look at the decision regions created by the Adaline model. We see that the line looks pretty optimal, and cuts about halfway between the two data regions.

```python
plot_decision_regions(X1_train, y1_train, classifier=ada1)
plt.title('Adaline - Gradient Descent')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Finally we use the model to predict the test set, and we find it predicts the test set with 100% accuracy.

```python
y1_ada_pred = ada1.predict(X1_test)
ada_misses1 = (y1_test != y1_ada_pred).sum()
ada_accuracy1 = (1 - ada_misses1/y1_ada_pred.size) * 100
print(f" model made {ada_misses1} misses out of {y1_ada_pred.size} tests for an accuracy of {ada_accuracy1}%")
```

# Linearly Inseparable Classes
## 2 Features
Now we will apply the models to a feature set where the data is linearly inseparable. We pick versicolor (rows 50-100) and virginica (rows 100-150) as our classes, and we'll use the features sepal length and petal width. We will set the target value to 1 for versicolor and -1 for virginica.

With the linearly separable classes, we didn't standardize the feature set. But we found that with the linearly inseparable classes, training was much improved if the data was standardized. So, we standardize the feature set, before dividing it into the training and test sets.

```python
# extract sepal length and petal length
#0=sepal length, 1=sepal width, 2=petal length, 3=petal width
X4 = df.iloc[50:150, [0, 3]].values

# standardize features
X4_std = np.copy(X4)
X4_std[:, 0] = (X4[:, 0] - X4[:, 0].mean()) / X4[:, 0].std()
X4_std[:, 1] = (X4[:, 1] - X4[:, 1].mean()) / X4[:, 1].std()

# select versicolor and virginica
y4 = df.iloc[50:150, 4].values
y4 = np.where(y4 == 'Iris-versicolor', 1, -1)

X4_train, X4_test, y4_train, y4_test = train_test_split(X4_std, y4, test_size=0.3, random_state=1, stratify=y4)
```

As before, we'll plot our data set to visualize it. This time we see that while there is a general separation between the data sets, there is some overlap, so the data classes are not linearly separable. Note that because we standardized our data, the mean value for each feature is 0, and the data is mostly contained with the range of -1 to 1.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# plot data
plt.scatter(X4_std[:50, 0], X4_std[:50, 1],
            color='red', marker='o', label='versicolor')
plt.scatter(X4_std[50:100, 0], X4_std[50:100, 1],
            color='blue', marker='x', label='virginica')

plt.xlabel('sepal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')

plt.show()
```

## 2 Features - Perceptron
Now we apply the Perceptron model to the linearly inseparable dataset. Unlike in the linearly separable case, we find that this time, the number of updates doesn't go to zero, even if we run the model for a lot of epochs. After 10 epochs, we find that it is making 9 updates per epoch toward the end of the training execution.

```python
ppn4 = Perceptron(eta=0.1, n_iter=10)

ppn4.fit(X4_train, y4_train, printEvery=1)

plt.plot(range(1, len(ppn4.errors_) + 1), ppn4.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')

plt.show()
```

Now we plot the decision regions generated by the model after 10 epochs. While Perceptron is not guaranteed to handle linearly inseparable data well, we find that the decision line actually looks pretty good for the given data.

```python
plot_decision_regions(X4_train, y4_train, classifier=ppn4)
plt.title('Perceptron')
plt.xlabel('sepal length')
plt.ylabel('petal width')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Now we test the model against the test set to find the accuracy. We find that the model predicts with 97% accuracy, which is extremely good, especially since the lack of linear separability means that 100% accuracy is not possible.

```python
y4_ppn_pred = ppn4.predict(X4_test)
ppn_misses4 = (y4_test != y4_ppn_pred).sum()
ppn_accuracy4 = (1 - ppn_misses4/y4_ppn_pred.size) * 100
print(f" model made {ppn_misses4} misses out of {y4_ppn_pred.size} tests for an accuracy of {ppn_accuracy4}%")
```

## 2 Features - Adaline
Now we train the Adaline model on the linearly inseparable 2 feature set. We now find that, since we are using standardized data, we are able to increase the eta to 0.01, which improves the learning time dramatically. We find that the model is already converging on a location after 10 epochs, as we can see by the fact that the sum-squared error decreases a lot after the first few epochs, but by the 10th epoch, the sum-squared error is hardly changing.

```python
ax = plt.axes()

ada4 = AdalineGD(n_iter=10, eta=0.01).fit(X4_train, y4_train, printEvery=1)
plt.plot(range(1, len(ada4.cost_) + 1), ada4.cost_, marker='o')
ax.set_xlabel('Epochs')
ax.set_ylabel('Sum-squared-error')
ax.set_title('Adaline')

plt.show()
```

Graphing the decision regions generated by the Adaline model, we find that it divides the data sets pretty well.

```python
plot_decision_regions(X4_train, y4_train, classifier=ada4)
plt.title('Adaline - Gradient Descent')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Finally we test the model against the test set, and find that it scores the same accuracy as the Perceptron model did.

```python
y4_ada_pred = ppn4.predict(X4_test)
ada_misses4 = (y4_test != y4_ada_pred).sum()
ada_accuracy4 = (1 - ada_misses4/y4_ada_pred.size) * 100
print(f" model made {ada_misses4} misses out of {y4_ada_pred.size} tests for an accuracy of {ada_accuracy4}%")
```

## 3 Features - Adaline
Now we apply the Adaline model to the data set. We again find that we can use eta=0.01, and the model levels off by about 10 epochs.

```python
ax = plt.axes()

ada5 = AdalineGD(n_iter=10, eta=0.01).fit(X5_train, y5_train, printEvery=1)
plt.plot(range(1, len(ada5.cost_) + 1), ada5.cost_, marker='o')
ax.set_xlabel('Epochs')
ax.set_ylabel('Sum-squared-error')
ax.set_title('Adaline')

plt.show()
```

Testing our Adaline model against the data set, we find that like the Perceptron model, it correct predicts 29 of the 30 records for a 97% accuracy.

```python
y5_ada_pred = ppn5.predict(X5_test)
ada_misses5 = (y5_test != y5_ada_pred).sum()
ada_accuracy5 = (1 - ada_misses5/y5_ada_pred.size) * 100
print(f" model made {ada_misses5} misses out of {y5_ada_pred.size} tests for an accuracy of {ada_accuracy5}%")
```

## 4 Features
Finally, we add the 4th feature to our data set for the linearly inseparable classes.

```python
# extract sepal length and petal length
#0=sepal length, 1=sepal width, 2=petal length, 3=petal width
X6 = df.iloc[50:150, [0, 1, 2, 3]].values

# standardize features
X6_std = np.copy(X6)
X6_std[:, 0] = (X6[:, 0] - X6[:, 0].mean()) / X6[:, 0].std()
X6_std[:, 1] = (X6[:, 1] - X6[:, 1].mean()) / X6[:, 1].std()
X6_std[:, 2] = (X6[:, 2] - X6[:, 2].mean()) / X6[:, 2].std()
X6_std[:, 3] = (X6[:, 3] - X6[:, 3].mean()) / X6[:, 3].std()

# select setosa and versicolor
y6 = df.iloc[50:150, 4].values
y6 = np.where(y6 == 'Iris-versicolor', 1, -1)

X6_train, X6_test, y6_train, y6_test = train_test_split(X6_std, y6, test_size=0.3, random_state=1, stratify=y6)
```

## 4 Features - Perceptron
We apply our Perceptron model to the 4-feature set, and find that it again oscillates at 4 to 8 updates per epoch.

```python
ppn6 = Perceptron(eta=0.1, n_iter=10)

ppn6.fit(X6_train, y6_train, printEvery=1)

plt.plot(range(1, len(ppn6.errors_) + 1), ppn6.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')

plt.show()
```

We apply our trained Perceptron model to the test set, and find that this time, it predicts all 30 test records correctly.

```python
y6_ppn_pred = ppn6.predict(X6_test)
ppn_misses6 = (y6_test != y6_ppn_pred).sum()
ppn_accuracy6 = (1 - ppn_misses6/y6_ppn_pred.size) * 100
print(f" model made {ppn_misses6} misses out of {y6_ppn_pred.size} tests for an accuracy of {ppn_accuracy6}%")
```

## 4 Features - Adaline
Now we train the Adaline model with the 4 feature data set. We find that using the same eta as before (0.01), the model spirals out of control. By trial and error, we find that we have to reduce the eta to 0.003. At that learning rate, the model trains quickly and begins to level off by 10 epochs.

```python
ax = plt.axes()

ada6 = AdalineGD(n_iter=10, eta=0.003).fit(X6_train, y6_train, printEvery=1)
plt.plot(range(1, len(ada6.cost_) + 1), ada6.cost_, marker='o')
ax.set_xlabel('Epochs')
ax.set_ylabel('Sum-squared-error')
ax.set_title('Adaline - Learning rate 0.0001')

plt.show()
```

Testing our trained Adaline model against the data set, we find that like the Perceptron model, it predicts all 30 records in the test set correctly.

```python
y6_ada_pred = ppn6.predict(X6_test)
ada_misses6 = (y6_test != y6_ada_pred).sum()
ada_accuracy6 = (1 - ada_misses6/y6_ada_pred.size) * 100
print(f" model made {ada_misses6} misses out of {y6_ada_pred.size} tests for an accuracy of {ada_accuracy6}%")
```

# Summary
In summary, both the Perceptron and Adaline models were very successful with the data set. Following the lead of the book, we looked at the updates per epoch for the Perceptron model, while looking at the sum-squared-error for the Adaline model. While the Perceptron model did continue making updates for the linearly-inseparable classes, it still generated good predictions after only 10 epochs; its accuracy was nearly perfect, and equivalent to the Adaline model. With other data sets, especially those with linearly inseparable classes, the Perceptron might not do as well, but on the data sets we chose, it did very well. In addition, the Adaline model was much more finicky. It trained very slowly on data that was not standardized, and we had to tweak the learning rate to find a balance between learning extremely slowly on the one hand, or spiraling out of control on the other. With other data sets, we would probably find that the Adaline model performs better; but on this training set, it was more finicky to deal with and didn't provide any improvement over the Perceptron model.
